**What is the difference between numpy and pandas?**

In numpy only with index value you can access the element

In pandas we can use both row name, column name and also index values in order to access the elements



```python
import pandas as pd
import numpy as np
```

```python
b = np.array([1,2,3,4,5])
b
```

```python
a = pd.Series([1,2,3,4,5,6],dtype=np.float64)
a
```

```python
print(type(a[0]))
print(type(a))
```

```python
a.drop(3)
```

In the above code the element is dropped but the index value of other numbers did not change, but in case of numpy there will be a shift in the index value that can be accessed


```python
a[3]
```

```python
a.drop(3,inplace=True)# only when we give inplace = True the element will be fully dropped
```

```python
a.iloc[3]
```

```python
a.loc[3]
```

iloc we access with the help of index no

in loc we access with the help of element
