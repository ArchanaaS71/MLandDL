```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df['Species']=lb.fit_transform(df['Species'])
```

```python
df
```

```python
x = df.drop('Species',axis=1)
y = df['Species']
```

```python
from sklearn.preprocessing import StandardScaler
sb = StandardScaler()
x=sb.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
learning_rate=[0.05,0.1,0.25,0.5,0.75,1]
for learning_rate in learning_rate:
  ab = AdaBoostClassifier(learning_rate=learning_rate,n_estimators=100,random_state=42)
  ab.fit(x_train,y_train)
  print("Learning Rate:",learning_rate)
  print("Accuracy score (Training): {0:.3f}".format(ab.score(x_train,y_train)))
  print("Accuracy score (Testing): {0:.3f}".format(ab.score(x_test,y_test)))
  print()
```
