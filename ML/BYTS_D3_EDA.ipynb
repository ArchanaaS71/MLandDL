Identifying Outliers:
- Z - Score: Z-score is also called standard score.How far does the point lie from the mean value.after setting up of threshold value one can utilize z-score values to help define the outliers.
- IQR(Inter Quartile Range:finding the outliers is the most commonly used and most trusted approach used in research field.


```python
from scipy import stats
hp_zscore = np.abs(stats.zscore(df['normalized-losses']))
hp_zscore
```

```python
df['normalized-losses']
```

```python
import seaborn as sns
sns.boxplot(x=hp_zscore)
```

```python
threshold = hp_zscore.max()
```

```python
index_val = np.where(hp_zscore >= threshold)
df.iloc[index_val]['normalized-losses']
```

```python
#drop the row that is an outlier
new_df = df.drop(index=df.iloc[index_val]['normalized-losses'].index)
new_df
```

```python
q1 = df['length'].quantile(0.25)
q3 = df['length'].quantile(0.75)
iqr = q3-q1
lower = q1 - 1.5*iqr
upper = q3 +1.5*iqr
```

```python
upper_array = np.where(df['length']>upper)
lower_array = np.where(df['length']<lower)
```

```python
print(upper_array)
print(lower_array)
```
